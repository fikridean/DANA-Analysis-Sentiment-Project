```python
from google_play_scraper import app, reviews, Sort, reviews_all
```

```python
scrapreview = reviews(
    'id.dana',          
    lang='en',             
    country='id',          
    sort=Sort.MOST_RELEVANT,
    count=10000             
)
```

```python
scrapreview[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame([review for review in scrapreview[0]])
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
df.info()
```

```python
new_df = df.drop(['userName', 'userImage', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt', 'appVersion'], axis=1)
```

```python
new_df.to_csv('ulasan_aplikasi.csv', index=False)
```
